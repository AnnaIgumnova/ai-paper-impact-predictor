```python
import pandas as pd
from pycaret.classification import load_model

```

```python
# Load the tuned Gradient Boosting Classifier model from PyCaret saved model file
tuned_gbc = load_model('../models/gbc_tuned')
```

```python
# Checking the steps of the loaded model pipeline
tuned_gbc.steps
```

The GBC model is the last step in the pipeline, so we need to extract this one

```python
import pickle as pkl
```

```python
final_gbc = tuned_gbc.named_steps['trained_model']
```

```python
with open('../models/gbc_tuned_streamlit.pkl', 'wb') as f:
    pkl.dump(final_gbc, f)
```

```python
import pickle as pkl
import pandas as pd

with open("../models/gbc_tuned_streamlit.pkl", "rb") as f:
    model = pkl.load(f)

X_test = pd.read_csv("../data/features/X_test_trimmed.csv")
probs = model.predict_proba(X_test)[:, 1]

print(pd.Series(probs).describe())
print(pd.Series(probs).quantile([0.25, 0.5, 0.75, 0.90, 0.95]))
```
