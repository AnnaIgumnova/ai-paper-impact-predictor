```python
import shap
import pandas as pd
import pickle

with open("../models/gbc_tuned_streamlit.pkl", "rb") as f:
    model = pkl.load(f)

X_train = pd.read_csv("../data/features/X_train_trimmed.csv")

# Create SHAP explainer using training data as background
explainer = shap.TreeExplainer(model, X_train)

# Test on first row of test set
X_test = pd.read_csv("../data/features/X_test_trimmed.csv")
shap_values = explainer(X_test.iloc[[0]], check_additivity=False)
print(shap_values)
```

```python
import matplotlib.pyplot as plt

shap.plots.waterfall(shap_values[0])
plt.tight_layout()
plt.show()
```
